Создать базовый класс Student в C#, который будет представлять информацию о
студентах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.
#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public class Student
{
    private string name;
    private string age;
    private string course;
    private string studentId;
    private double gpa;
    private bool isActive;

    public string Name
    {
        get { return name; }
        set { name = value; }
    }
    public string Age
    {
        get { return age; }
        set { age = value; }
    }
    public string Course
    {
        get { return course; }
        set { course = value; }
    }
    public string StudentId
    {
        get { return studentId; }
        set { studentId = value; }
    }
    public double GPA
    {
        get { return gpa; }
        set { gpa = value; }
    }
    public bool IsActive
    {
        get { return isActive; }
        set { isActive = value; }
    }

    public Student(string name, string age, string course, string studentId, double gpa = 0.0)
    {
        Name = name;
        Age = age;
        Course = course;
        StudentId = studentId;
        GPA = gpa;
        IsActive = true;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}; \nВозраст: {Age}; \nКурс: {Course}; \nID: {StudentId}; \nGPA: {GPA:F2}; \nАктивен: {(IsActive ? "Да" : "Нет")}";
    }
    
    public virtual void Study()
    {
        Console.WriteLine($"{Name} изучает предметы.");
    }
    
    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдает экзамен.");
    }
    
    public virtual void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} общается с {otherStudent.Name}.");
    }
    
    // Новые методы
    public virtual void CalculateGPA(double newGrade)
    {
        GPA = (GPA + newGrade) / 2;
        Console.WriteLine($"{Name} получил оценку {newGrade:F1}. Новый GPA: {GPA:F2}");
    }
    
    public void ChangeStatus(bool status)
    {
        IsActive = status;
        Console.WriteLine($"{Name} статус изменен на: {(status ? "Активен" : "Неактивен")}");
    }
    
    public virtual void AttendLecture()
    {
        Console.WriteLine($"{Name} посещает лекцию.");
    }
}

// Интерфейсы для множественного наследования
public interface IResearchable
{
    void PublishPaper();
    void AttendConference();
}

public interface IScholarshipEligible
{
    void ApplyForScholarship();
    bool CheckScholarshipEligibility();
}

public interface ITeachable
{
    void ConductSeminar();
    void MentorStudents();
}

public class BachelorStudent : Student, IScholarshipEligible
{
    private string specialty;
    private int projectsCompleted;
    private string internshipCompany;
    
    public string Specialty
    {
        get { return specialty; }
        set { specialty = value; }
    }
    
    public int ProjectsCompleted
    {
        get { return projectsCompleted; }
        set { projectsCompleted = value; }
    }
    
    public string InternshipCompany
    {
        get { return internshipCompany; }
        set { internshipCompany = value; }
    }
    
    public BachelorStudent(string name, string age, string course, string studentId, string specialty, double gpa = 0.0)
        : base(name, age, course, studentId, gpa)
    {
        Specialty = specialty;
        ProjectsCompleted = 0;
        InternshipCompany = "Не назначена";
    }
    
    public override string GetInfo()
    {
        return base.GetInfo() + $"; \nСпециальность: {Specialty}; \nЗавершенные проекты: {ProjectsCompleted}; \nСтажировка: {InternshipCompany}";
    }
    
    public override void Study()
    {
        Console.WriteLine($"{Name} (бакалавр, специальность: {Specialty}) изучает профильные предметы.");
    }

    public void DoProject()
    {
        ProjectsCompleted++;
        Console.WriteLine($"{Name} (бакалавр, специальность: {Specialty}) работает над проектом. Всего завершено: {ProjectsCompleted}");
    }
    
    public void StartInternship(string company)
    {
        InternshipCompany = company;
        Console.WriteLine($"{Name} начал стажировку в компании: {company}");
    }
    
    public override void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} (бакалавр) обменивается опытом с {otherStudent.Name}.");
    }
    
    // Реализация интерфейса IScholarshipEligible
    public void ApplyForScholarship()
    {
        if (GPA >= 3.5)
        {
            Console.WriteLine($"{Name} подает заявку на стипендию. Отличный GPA!");
        }
        else
        {
            Console.WriteLine($"{Name} GPA слишком низкий для стипендии.");
        }
    }
    
    public bool CheckScholarshipEligibility()
    {
        return GPA >= 3.5 && IsActive && ProjectsCompleted >= 2;
    }
}

public class MasterStudent : Student, IResearchable, ITeachable
{
    private string scientificAdvisor;
    private string researchTopic;
    private int papersPublished;
    
    public string ScientificAdvisor
    {
        get { return scientificAdvisor; }
        set { scientificAdvisor = value; }
    }
    
    public string ResearchTopic
    {
        get { return researchTopic; }
        set { researchTopic = value; }
    }
    
    public int PapersPublished
    {
        get { return papersPublished; }
        set { papersPublished = value; }
    }
    
    public MasterStudent(string name, string age, string course, string studentId, string advisor, string researchTopic, double gpa = 0.0)
        : base(name, age, course, studentId, gpa)
    {
        ScientificAdvisor = advisor;
        ResearchTopic = researchTopic;
        PapersPublished = 0;
    }
    
    public override string GetInfo()
    {
        return base.GetInfo() + $"; \nНаучный руководитель: {ScientificAdvisor}; \nТема исследования: {ResearchTopic}; \nОпубликовано статей: {PapersPublished}";
    }
    
    public override void TakeExam()
    {
        Console.WriteLine($"{Name} (магистр) сдает экзамен под наблюдением научного руководителя {ScientificAdvisor}.");
    }

    public void DoResearch()
    {
        Console.WriteLine($"{Name} (магистр) проводит научное исследование по теме: {ResearchTopic}.");
    }
    
    public override void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} (магистр) консультирует {otherStudent.Name}.");
    }
    
    // Реализация интерфейса IResearchable
    public void PublishPaper()
    {
        PapersPublished++;
        Console.WriteLine($"{Name} опубликовал научную статью. Всего публикаций: {PapersPublished}");
    }
    
    public void AttendConference()
    {
        Console.WriteLine($"{Name} участвует в научной конференции с докладом по теме: {ResearchTopic}");
    }
    
    // Реализация интерфейса ITeachable
    public void ConductSeminar()
    {
        Console.WriteLine($"{Name} проводит семинар для студентов младших курсов.");
    }
    
    public void MentorStudents()
    {
        Console.WriteLine($"{Name} выступает в роли ментора для студентов-бакалавров.");
    }
}

public class PhDStudent : Student, IResearchable, ITeachable, IScholarshipEligible
{
    private string dissertationTopic;
    private string researchDepartment;
    private int yearsToGraduation;
    private bool hasFunding;
    
    public string DissertationTopic
    {
        get { return dissertationTopic; }
        set { dissertationTopic = value; }
    }
    
    public string ResearchDepartment
    {
        get { return researchDepartment; }
        set { researchDepartment = value; }
    }
    
    public int YearsToGraduation
    {
        get { return yearsToGraduation; }
        set { yearsToGraduation = value; }
    }
    
    public bool HasFunding
    {
        get { return hasFunding; }
        set { hasFunding = value; }
    }
    
    public PhDStudent(string name, string age, string course, string studentId, string topic, string department, int years, bool funding, double gpa = 0.0)
        : base(name, age, course, studentId, gpa)
    {
        DissertationTopic = topic;
        ResearchDepartment = department;
        YearsToGraduation = years;
        HasFunding = funding;
    }
    
    public override string GetInfo()
    {
        return base.GetInfo() + $"; \nТема диссертации: {DissertationTopic}; \nКафедра: {ResearchDepartment}; \nЛет до выпуска: {YearsToGraduation}; \nФинансирование: {(HasFunding ? "Есть" : "Нет")}";
    }
    
    public override void Study()
    {
        Console.WriteLine($"{Name} (аспирант) изучает материалы, связанные с темой: {DissertationTopic}.");
    }

    public void WriteDissertation()
    {
        Console.WriteLine($"{Name} (аспирант) пишет диссертацию на тему: {DissertationTopic}.");
    }
    
    public void DefendProposal()
    {
        Console.WriteLine($"{Name} защищает предложение по диссертации перед комиссией.");
    }
    
    public override void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} (аспирант) делится результатами исследований с {otherStudent.Name}.");
    }
    
    // Реализация интерфейса IResearchable
    public void PublishPaper()
    {
        Console.WriteLine($"{Name} публикует статью в рецензируемом журнале по теме: {DissertationTopic}");
    }
    
    public void AttendConference()
    {
        Console.WriteLine($"{Name} представляет исследование на международной конференции.");
    }
    
    // Реализация интерфейса ITeachable
    public void ConductSeminar()
    {
        Console.WriteLine($"{Name} проводит продвинутый семинар для магистрантов.");
    }
    
    public void MentorStudents()
    {
        Console.WriteLine($"{Name} руководит научной работой магистрантов.");
    }
    
    // Реализация интерфейса IScholarshipEligible
    public void ApplyForScholarship()
    {
        if (!HasFunding)
        {
            Console.WriteLine($"{Name} подает заявку на исследовательский грант.");
        }
        else
        {
            Console.WriteLine($"{Name} уже имеет финансирование.");
        }
    }
    
    public bool CheckScholarshipEligibility()
    {
        return GPA >= 3.7 && IsActive && !HasFunding;
    }
}

// Демонстрация работы расширенной системы
Student[] students = new Student[]
{
    new BachelorStudent("Алмаз", "19", "2", "B2024001", "Искусственный интеллект и программирование", 3.8),
    new MasterStudent("Иван", "24", "1", "M2024001", "проф. Пупкин", "Машинное обучение", 3.9),
    new PhDStudent("Олеся", "28", "3", "P2024001", "Оптимизация распределенных систем", "Кафедра информатики", 2, true, 4.0)
};

// Демонстрация базовых методов
foreach (var s in students)
    {
    Console.WriteLine("=== ИНФОРМАЦИЯ О СТУДЕНТЕ ===");
    Console.WriteLine(s.GetInfo());
    Console.WriteLine("\n=== ДЕЯТЕЛЬНОСТЬ ===");
    s.Study();
    s.TakeExam();
    s.AttendLecture();
    s.CalculateGPA(4.2);
    Console.WriteLine();
}

// Демонстрация взаимодействия
Console.WriteLine("=== ВЗАИМОДЕЙСТВИЕ ===");
students[0].InteractWith(students[1]);
students[1].InteractWith(students[2]);
students[2].InteractWith(students[0]);

// Демонстрация специфических методов
Console.WriteLine("\n=== СПЕЦИФИЧЕСКИЕ МЕТОДЫ ===");
((BachelorStudent)students[0]).DoProject();
((BachelorStudent)students[0]).StartInternship("TechCorp");

((MasterStudent)students[1]).DoResearch();
((MasterStudent)students[1]).PublishPaper();

((PhDStudent)students[2]).WriteDissertation();
((PhDStudent)students[2]).DefendProposal();

// Демонстрация множественного наследования через интерфейсы
Console.WriteLine("\n=== МНОЖЕСТВЕННОЕ НАСЛЕДОВАНИЕ (ИНТЕРФЕЙСЫ) ===");

// интерфейс IResearchable
Console.WriteLine("--- Исследовательская деятельность ---");
((IResearchable)students[1]).AttendConference();
((IResearchable)students[2]).PublishPaper();

// интерфейс ITeachable
Console.WriteLine("--- Преподавательская деятельность ---");
((ITeachable)students[1]).ConductSeminar();
((ITeachable)students[2]).MentorStudents();

// интерфейс IScholarshipEligible
Console.WriteLine("--- Стипендии и гранты ---");
((IScholarshipEligible)students[0]).ApplyForScholarship();
Console.WriteLine($"Стипендия доступна: {((IScholarshipEligible)students[0]).CheckScholarshipEligibility()}");

((IScholarshipEligible)students[2]).ApplyForScholarship();
```
