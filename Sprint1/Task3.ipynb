Создать базовый класс Student в C#, который будет представлять информацию о
студентах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

#### Дополнительное задание
Добавьте к сущестующим классам конструктора классов с использованием гетторов и сетторов и реализуйте взаимодействие объектов между собой

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public class Student
{
    private string name;
    private string age;
    private string course;

    public string Name
    {
        get { return name; }
        set { name = value; }
    }
    public string Age
    {
        get { return age; }
        set { age = value; }
    }
    public string Course
    {
        get { return course; }
        set { course = value; }
    }

    public Student(string name, string age, string course)
    {
        Name = name;
        Age = age;
        Course = course;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}; \nВозраст: {Age}; \nКурс: {Course}";
    }
    public virtual void Study()
    {
        Console.WriteLine($"{Name} изучает предметы.");
    }
    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдает экзамен.");
    }
    public virtual void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} общается с {otherStudent.Name}.");
    }
}

public class BachelorStudent : Student
{
    private string specialty;
    public string Specialty
    {
        get { return specialty; }
        set { specialty = value; }
    }
    public BachelorStudent(string name, string age, string course, string specialty)
        : base(name, age, course)
    {
        Specialty = specialty;
    }
    public override string GetInfo()
    {
        return base.GetInfo() + $"; \nСпециальность: {Specialty}.";
    }
    public override void Study()
    {
        Console.WriteLine($"{Name} (бакалавр, специальность: {Specialty} ) изучает профильные предметы.");
    }

    public void DoProject()
    {
        Console.WriteLine($"{Name} (бакалавр, специальность: {Specialty}) работает над проектом.");
    }
    public override void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} (бакалавр) обменивается опытом с {otherStudent.Name}.");
    }
}

public class MasterStudent : Student
{
    private string scientificAdvisor;
    public string ScientificAdvisor
    {
        get { return scientificAdvisor; }
        set { scientificAdvisor = value; }
    }
    public MasterStudent(string name, string age, string course, string advisor)
        : base(name, age, course)
    {
        ScientificAdvisor = advisor;
    }
    public override string GetInfo()
    {
        return base.GetInfo() + $"; \nНаучный руководитель: {ScientificAdvisor}.";
    }
    public override void TakeExam()
    {Console.WriteLine($"{Name} (магистр) сдает экзамен под наблюдением научного руководителя {ScientificAdvisor}.");
    }

    public void DoResearch()
    {
        Console.WriteLine($"{Name} (магистр) проводит научное исследование.");
    }
    public override void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} (магистр) консультирует {otherStudent.Name}.");
    }
}

public class PhDStudent : Student
{
    private string dissertationTopic;
    public string DissertationTopic
    {
        get { return dissertationTopic; }
        set { dissertationTopic = value; }
    }
    public PhDStudent(string name, string age, string course, string topic)
        : base(name, age, course)
    {
        DissertationTopic = topic;
    }
    public override string GetInfo()
    {
        return base.GetInfo() + $"; Тема диссертации: {DissertationTopic}.";
    }
    public override void Study()
    {
        Console.WriteLine($"{Name} (аспирант) изучает материалы, связанные с темой: {DissertationTopic}.");
    }

    public void WriteDissertation()
    {
        Console.WriteLine($"{Name} (аспирант) пишет диссертацию на тему: {DissertationTopic}.");
    }
    public override void InteractWith(Student otherStudent)
    {
        Console.WriteLine($"{Name} (аспирант) делится результатами исследований с {otherStudent.Name}.");
    }
}
Student[] students = new Student[]
    {
        new BachelorStudent("Алмаз", "19", "2", "Искусственный интеллект и программирование"),
        new MasterStudent("Иван", "24", "1", "проф.Пупкин"),
        new PhDStudent("Олеся", "28", "3", "Оптимизация распределенных систем")
    };
foreach (var s in students)
    {
        Console.WriteLine(s.GetInfo());
        s.Study();
        s.TakeExam();
        Console.WriteLine();
    }
        students[0].InteractWith(students[1]);
        students[1].InteractWith(students[2]);
        students[2].InteractWith(students[0]);
        ((BachelorStudent)students[0]).DoProject();
        ((MasterStudent)students[1]).DoResearch();
        ((PhDStudent)students[2]).WriteDissertation();
```
